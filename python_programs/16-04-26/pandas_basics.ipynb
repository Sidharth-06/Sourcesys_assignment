## 10. Real dataset: Iris

This section uses the Iris CSV stored in the workspace to demonstrate common pandas operations on a real dataset.

```python
import pandas as pd
from pathlib import Path
from IPython.display import display

candidate_paths = [
    Path('python_programs/15-04-26/data/iris.csv'),
    Path('../15-04-26/data/iris.csv'),
    Path('../../python_programs/15-04-26/data/iris.csv'),
]

iris_path = None
for candidate in candidate_paths:
    if candidate.exists():
        iris_path = candidate
        break

if iris_path is None:
    raise FileNotFoundError('Could not locate iris.csv in the workspace.')

iris = pd.read_csv(
    iris_path,
    header=None,
    names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
)

print(f'Loaded dataset from: {iris_path}')
print('Shape:', iris.shape)
print('Columns:', list(iris.columns))

print('\nHead:')
display(iris.head())

print('\nTail:')
display(iris.tail())

print('\nData types:')
display(iris.dtypes.to_frame('dtype'))

print('\nMissing values:')
display(iris.isna().sum().to_frame('missing_count'))

print('\nDescriptive statistics:')
display(iris.describe())

print('\nSpecies counts:')
display(iris['species'].value_counts())

print('\nAverage measurements by species:')
display(iris.groupby('species').mean(numeric_only=True))

print('\nTop 5 rows by petal length:')
display(iris.sort_values('petal_length', ascending=False).head())

iris['sepal_area'] = iris['sepal_length'] * iris['sepal_width']

print('\nTop 5 rows by sepal area:')
display(iris.nlargest(5, 'sepal_area'))

print('\nCorrelation matrix for numeric columns:')
display(iris.select_dtypes(include='number').corr())

print('\nRandom sample:')
display(iris.sample(5, random_state=42))
```
