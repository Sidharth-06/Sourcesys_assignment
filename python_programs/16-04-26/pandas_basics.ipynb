# Pandas Basics Exploration



```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 100)
```

## 1. Create Series and DataFrames


```python
scores = pd.Series([88, 91, 76, 95], index=['Asha', 'Ben', 'Chirag', 'Diya'])
scores
```

```python
student_data = {
    'name': ['Asha', 'Ben', 'Chirag', 'Diya', 'Esha'],
    'age': [20, 21, 19, 22, 20],
    'marks': [88, 91, 76, 95, np.nan],
    'city': ['Delhi', 'Mumbai', 'Delhi', 'Chennai', 'Delhi']
}

df = pd.DataFrame(student_data)
df
```

## 4. Clean missing values



```python
df.isna()
```

```python
df.fillna({'marks': df['marks'].mean()})
```

```python
df.dropna()
```

## 5. Sort, rename, and transform



```python
df.sort_values(by='marks', ascending=False)
```

```python
df.rename(columns={'marks': 'score'})
```

```python
df.drop(columns=['city'])
```

```python
df['marks_plus_5'] = df['marks'] + 5
df['passed'] = df['marks'] >= 80
df
```

```python
df['city'].value_counts()
```

## 6. Group by and aggregate


```python
df.groupby('city')['marks'].mean()
```

```python
df.groupby('city').agg({'marks': ['mean', 'max', 'min'], 'age': 'count'})
```

## 7. Combine DataFrames


```python
left = pd.DataFrame({'student_id': [1, 2, 3], 'name': ['Asha', 'Ben', 'Chirag']})
right = pd.DataFrame({'student_id': [1, 2, 3], 'course': ['Python', 'SQL', 'Pandas']})

pd.merge(left, right, on='student_id')
```

```python
first_half = df.head(2)
second_half = df.tail(2)
pd.concat([first_half, second_half], ignore_index=True)
```
